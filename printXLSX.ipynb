```python
import xpress as xp

# 定义模型
prob = xp.problem()

# 添加变量、约束、目标函数（示例）
x = xp.var(name="x", lb=0)
y = xp.var(name="y", lb=0)
prob.addVariable(x, y)
prob.addConstraint(x >= 0)
prob.addConstraint(y >= 0)
prob.addConstraint(x + y <= 10)
prob.setObjective(x + 2*y, sense=xp.maximize)

# 设置求解参数
prob.setControl("MAXTIME", 30)  # 5分钟
prob.setControl("MAXMIPSOL", 1000000)  # 允许存储多个可行解

# 定义回调函数记录找到的解
feasible_solutions = []

# 定义回调函数
def my_callback(prob):
    solution_values = prob.getSolution([x, y])  # 获取变量值
    objval = prob.getObjVal()  # 获取目标值
    feasible_solutions.append((solution_values, objval))
    print(f"找到可行解: x={solution_values[0]}, y={solution_values[1]}, 目标值={objval}")

# 绑定整数解回调
prob.addcbintsol(my_callback)


# prob.setControl("MIPDUALREDUCTIONS", 0)   # 禁用对偶缩减，避免遗漏可行解
# prob.setControl("MAXMIPSOL", 1000)         # 最多找到1000个可行解
# 求解
prob.solve()
print(f"最优解: {prob.getObjVal()}")
# 输出x y
print("solution:", prob.getSolution())

# 输出找到的可行解数量
num_solutions = prob.getAttrib("SOLVEDOBJS")
print(f"找到的可行解总数: {num_solutions}") 
```
